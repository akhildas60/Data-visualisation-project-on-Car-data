```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)
```

```python
#drop the nulls
df.dropna()
```

```python
df['Brand'] = df['Brand'].str.replace('vokswagen','volkswagen')
df['Brand'] = df['Brand'].str.replace('chevroelt','chevrolet')
df['Brand'] = df['Brand'].str.replace('mercedes-benz','mercedes')
df['Brand'] = df['Brand'].str.replace('toyouta','toyota')
df['Brand'] = df['Brand'].str.replace('chevy','chevrolet')
df.head()
```

```python
print(df['model_year'].unique())
```

```python
plt.figure(figsize =(25,7))
ax = sns.countplot(data = df, x= "Brand")
plt.title("Brand distribution")
ax.bar_label(ax.containers[0])
plt.show()
#Here we can see that Ford is the most populer brand, chevrolet is also populer
```

```python
plt.figure(figsize = (5,5))
ax=sns.countplot(data=df, x= "origin")
plt.title("Brand's origin distribution ")
ax.bar_label(ax.containers[0])
plt.show()
#Usa dominates here by a very large margin
```

```python
selected_variables = ['mpg', 'displacement', 'cylinders', 'weight', 'horsepower', 'acceleration']
selected_df = df[selected_variables]

# Calculating the correlation matrix
correlation_matrix = selected_df.corr()

# Displaying the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Correlation Matrix with Color")
plt.show()
```

```python
# From this correlation matrix we can get the relationship between the four variables,
#Diplacement and cylinders are highly positively related
#similery weight and milage is negatively related hence if our focus is milage then less weighed car is preffered
```
